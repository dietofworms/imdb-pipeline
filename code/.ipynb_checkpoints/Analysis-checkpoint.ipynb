Now that we have a nice, clean `.CSV` file, we can analyse and visualise the data. We will use packages like `pandas`, `numpy`, `matplotlib` and more. 

```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline
```

### Load and clean the data frame
After processing, we can now read in the previously created `.CSV` file. In the import, we specify `index_col = 0` to use the outputted index in the new data frame.

```python
df = pd.read_csv('../data/df.csv', index_col = 0)
df.head() # peak at the dataset
```

Now that the data is loaded into a pandas dataframe and we've had a peak using `head()`, we will ensure that the data types are mapped correctly.

```python
df.dtypes
```

Nearly all of the columns/variables mapped correctly except for `votes`. This is due to the `,` in the values to seperate thousands. We will now find, replace and convert the column to an `int`.

```python
# replace , values in count of votes
df['votes'] = df['votes'].str.replace(',','').astype(int)
```

### Exploratory Data Analysis
The data frame is in a workable form so now it is time to visualise it to explore the relationship between variables.

```python
df.describe()
```

#### Variable Distributions
We have two numerical variables. To understand them deeper than simple summary statistics, we will plot them as histograms.

```python
bins = 10
plt.hist(x = df['rating'], bins = bins, facecolor = 'blue', alpha = 0.5)
plt.title('Distribution of Rating')
plt.xlim(6, 9)
plt.show()
```

```python
bins = 10
plt.hist(x = df['votes'], bins = bins, facecolor = 'blue', alpha = 0.5)
plt.title('Distribution of Votes')
#plt.xlim(6, 9)
plt.show()
```

```python
plt.subplot(1,2,1)
sns.distplot(df['votes'],  bins = 10, color = 'r')

plt.subplot(1,2,2)
sns.distplot(df['rating'], bins = 10, color = 'g')

plt.show()
```

```python
sns.pairplot(df)
```

There is an interesting relationship between ratings and votes. Now we will explore that deeper.

#### Ratings / Votes
Now we will build a scatterplot to visualise the relationship between two variables where:
$$x = rating, y = votes$$

```python
plt.figure(figsize=(8,6))
ax = sns.scatterplot(x = df['rating'], y = df['votes'], 
                     s = 100,
                     hue = df['title'], 
                     legend = 'full',
                     data = df)
#plt.rcParams["figure.figsize"] = (10, 8)
plt.title('Film Scores and Votes')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

The scatterplot shows that movies with more votes seem to have a higher rating. This could be because better movies become more popular thus drawing in more votes or vice-versa.

#### ESRB Rated (G, PG, etc.)
Another interesting relationship is the film ESRB rating (G, PG, M, ... ) and overall score.

```python
rated_mean = pd.DataFrame(df.groupby(['rated'])['rating'].mean())
rated_mean.reset_index(level=0, inplace=True)
rated_mean = rated_mean.sort_values('rating')

bar = sns.barplot(x = rated_mean['rated'], y = rated_mean['rating'], 
                  data = rated_mean, palette = 'Blues_d')

plt.show()
```
